```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_ollama import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import PromptTemplate
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
from youtube_transcript_api._errors import TranscriptsDisabled

video_id = "4eMoZInoxCU"

try:
    api = YouTubeTranscriptApi()
    transcript_list = api.fetch(video_id, languages=["en", "hi"])
    transcript = " ".join(chunk.text for chunk in transcript_list)
    print(transcript)

except TranscriptsDisabled:
    print("Nothing Found")
```

```python
transcript_list
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size = 400, chunk_overlap = 100)
chunks = splitter.create_documents([transcript])
```

```python
len(chunks)
```

```python
chunks[0]
```

```python
embeddings = OllamaEmbeddings(model='llama3.2')
vector_store = FAISS.from_documents(chunks, embeddings)
```

```python
vector_store.index_to_docstore_id
```

```python
vector_store.get_by_ids(['cb550ee5-72d7-4584-ac06-6556b2f6e458'])
```

```python
retriever = vector_store.as_retriever(search_type = "similarity", search_kwargs={"K" : 4})
```

```python
from langchain_ollama import ChatOllama
llm = ChatOllama(
    model='llama3.2'
)
```

```python
prompt = PromptTemplate(
    template="""
    You're a helpful assistant
    Answer only from the provided transcript context.
    If the context is insufficient, just say you don't know.

    {context}
    Question : {question}
    """,
    input_variables=['context', 'question']
)
```

```python
question = 'is the topic of any Fort is there??'
retrieved_docs = retriever.invoke(question)
```

```python
context_text = "\n\n".join(docs.page_content for docs in retrieved_docs)
context_text
```

```python
final_prompt = prompt.invoke({
    "context" : context_text,
    "question" : question
})
```

```python
answer = llm.invoke(final_prompt)
print(answer.content)
```

```python
from langchain_core.runnables import RunnableParallel, RunnableLambda, RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
```

```python
def format_docs(retrieved_docs):
    context_text = "\n\n".join(docs.page_content for docs in retrieved_docs)
    return context_text
```

```python
parallel_chain = RunnableParallel({
    'context' : retriever | RunnableLambda(format_docs),
    'question' : RunnablePassthrough()
})
```

```python
parallel_chain.invoke("Tell about Bhagard")
```

```python
parser = StrOutputParser()
```

```python
main_chain = parallel_chain | prompt | llm | parser
```

```python
main_chain.invoke("Can You summarize the video and tell me about Bhagard!!")
```
